```python
import matplotlib.pyplot as plt
```

# Bar Charts

Create a bar chart for underground trainlines

```python
tube_lines = ["Bakerloo", "Central", "Circle", "District", "Hammersmith", "Jubilee", "Metropolitan", "Northern", "Piccadilly", "Victoria", "Waterloo"]

number_stops = [25, 49, 36, 60, 29, 27, 34, 52, 53, 16,2]

col=["brown","red", "yellow", "green", "pink","gray","purple","black","lightblue", "blue", "darkblue"]

plt.figure(figsize=(20,9))
plt.xlabel("Trainlines")
plt.ylabel("Number")
plt.title("Number of stops for each Trainline")
plt.xticks(rotation=45)
plt.bar(tube_lines, number_stops, color= col)


```

Create a bar chart showing the total viewwing hours of the most popilar Netflix series

```python
Netflix_Title = ["Bridgerton: Series 1", "Money Heist: Part 4", "Stranger Things: Season 3", "The Witcher: Season 1", "13 Reasons Why: Season 2", "13 Reasons Why: Season 1", "You: Season 2", "Stranger Things: Season 2","Money Heist: Part 3","Ginny & Georgia: Season 1"]

Viewing_Hours = [625, 619, 582, 541, 496, 476, 457, 427, 426, 381]
plt.figure(figsize=(20,9))
plt.xlabel("Netflix Titles")
plt.ylabel("Viewing Hours ('000,000')")
plt.title("Viewing Hours for Netflix Shows by Series")
plt.xticks(rotation=45)
plt.bar(Netflix_Title, Viewing_Hours, color="red")
```

# Line Charts

Combing Charts

```python
month = ["Jan","Feb","Mar","Apr","May","Jun","Jul","Aug","Sept","Oct","Nov","Dec"]
Manchester_rainfall = [2.44,2.09,1.71,1.65,1.71,2.01,2.32,2.44,2.44,2.81,2.72,2.69, ]
London_rainfall = [2.44,1.76,1.46,1.42,1.71,1.65,1.83,2.44,2.13,2.93,2.83,2.56, ]

plt.plot(month,Manchester_rainfall,label="Manchester")
plt.plot(month,London_rainfall,label="London")

plt.title("Average Rainfall in cm\nManchester & London")
plt.xlabel("Months")
plt.ylabel("Rainfall (cm)")
plt.legend()

plt.show()
```

Exercise

```python
month=["Jan","Feb","Mar", "Apr", "May", "Jun"]
hats=[5200, 4100, 6000, 6900, 6050, 7000]
gloves=[2100, 1050, 2950, 5000, 6300, 5200]

plt.plot(month,hats,label="Monthly Hat Sales")
plt.plot(month,gloves,label="Monthly Glove Sales")

plt.title("Monthly sales for hats and gloves")
plt.xlabel("Months")
plt.ylabel("Sales")
plt.legend()

plt.show()
```
